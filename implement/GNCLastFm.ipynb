```python
###################################################################
#bc approx on lastfm
edges = []
nodes = set()
with open('C:/Users/pc/Desktop/thesis/data/dataLastFm/data.csv', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)

for datum in data:
    edges.append(datum)
    nodes.add(datum[0])
    nodes.add(datum[1])
network = nx.Graph()
network.add_nodes_from(nodes)
network.add_edges_from(edges)
```

```python
###################################################################
#revised on lastfm dataset
start_time1 = time.time()
for i in range(10):
    result = revised_girvan(network)
communities = next(result)
print(str((time.time() - start_time1)) + " seconds")
```

```python
draw_community(network, communities)
```
